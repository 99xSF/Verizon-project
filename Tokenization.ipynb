```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

filepath = '/content/drive/My Drive/processed_data.csv'
data = pd.read_csv(filepath)
data
```

```python
import nltk
from nltk.tokenize import word_tokenize

# Download the necessary NLTK resources
nltk.download('punkt')

# Tokenize and lowercase the text
def tokenize_lowercase(text):
    tokens = word_tokenize(text)  # Tokenize the text
    tokens = [word.lower() for word in tokens]  # Convert to lowercase
    return tokens

data['tokens'] = data['processed_description'].apply(tokenize_lowercase)

# Display tokenized data
print(data[['Product Name', 'tokens']].head())
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Lemmatize the tokens
def lemmatize_words(tokens):
    return [lemmatizer.lemmatize(token) for token in tokens]

data['lemmatized_tokens'] = data['tokens'].apply(lemmatize_words)

# Display lemmatized tokens
print(data[['Product Name', 'lemmatized_tokens']].head())
```

```python
import re

# Function to remove punctuation from tokens
def remove_punctuation(tokens):
    tokens = [re.sub(r'[^\w\s]', '', token) for token in tokens]  # Remove punctuation
    return [token for token in tokens if token]  # Remove empty tokens

data['cleaned_tokens'] = data['lemmatized_tokens'].apply(remove_punctuation)

# Display cleaned tokens
print(data[['Product Name', 'cleaned_tokens']].head())
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

# Define stopwords
stop_words = set(stopwords.words('english'))

# Remove stopwords from tokens
def remove_stopwords(tokens):
    return [token for token in tokens if token not in stop_words]

data['final_tokens'] = data['cleaned_tokens'].apply(remove_stopwords)

# Display final tokens after stopword removal
print(data[['Product Name', 'final_tokens']].head())
```

```python
# Function to normalize text (expand common contractions)
contraction_map = {"can't": "cannot", "won't": "will not", "n't": "not", "'re": "are", "'s": "is", "'d": "would", "'ll": "will"}

def normalize_text(tokens):
    return [contraction_map.get(token, token) for token in tokens]

data['normalized_tokens'] = data['final_tokens'].apply(normalize_text)

# Display normalized tokens
print(data[['Product Name', 'normalized_tokens']].head())
```

```python
# Convert tokens back into a single string
data['final_text'] = data['normalized_tokens'].apply(lambda tokens: ' '.join(tokens))

# Save the final dataset
data[['Product Name', 'final_text']].to_csv('final_processed_data.csv', index=False)

# Display the final processed data
# print(data[['Product Name', 'final_text']].head())
```
